```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql
```

```python
pymysql.install_as_MySQLdb()
```

# Main Harry Potter Character File Upload

```python
character_main = "characters_main.csv"
character_main_df = pd.read_csv(character_main)
character_main_df.head()
```

# Attribute Harry Potter Character File Upload 

```python
character_attributes = "character_attr.csv"
character_attributes_df = pd.read_csv(character_attributes)
character_attributes_df.head()
```

# Aggression Harry Potter Character File Upload

```python
character_aggression = "harry_potter_aggression_full.csv"
character_aggression_df = pd.read_csv(character_aggression)
character_aggression_df.head()
```
